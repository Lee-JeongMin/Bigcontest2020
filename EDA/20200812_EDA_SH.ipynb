```python
# 전체 topic list. 중복제거 시 1403개에서 939개로 축소
topiclist= []
for lists in ws_topic_data['TOPIC_WORD_LIST']:
    for topic in lists:
        topiclist.append(topic)
```

```python
pd.set_option('display.max_rows', 10000)

pd.Series(topiclist).value_counts().sort_values(ascending= False)
```

```python
# 동별 토픽리스트들 dictionary
dong2topiclist= dict(np.array(ws_topic_data[['DONG_NM','TOPIC_WORD_LIST']]))
```

```python
# 토픽이 속한 동 dictionary
topic2dong= {}

for dong, topics in dong2topiclist.items():
    for topic in topics:
        if topic in topiclist:
            topic2dong.setdefault(topic, dong)
        else:
            continue
```
